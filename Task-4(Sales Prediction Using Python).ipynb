```python
x=task_Advertising[['TV']]
y=task_Advertising[['Sales']]
```

```python
x_train, X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
ML = LinearRegression()
ML.fit(x_train,y_train)
```

```python
x_train_prediction=ML.predict(x_train)
print(x_train_prediction)
```

```python
print(y_train)
```

```python
print(ML.coef_)
```

```python
print(ML.intercept_)
```

```python
x_test_prediction=ML.predict(x_train)
print(x_test_prediction)
```

```python
plt.figure(figsize=(9,6))
plt.scatter(y_train ,x_train_prediction)
plt.xlabel('original values')
plt.ylabel('predicted values')
plt.title("Orginal Values and Predicted Values")
```
